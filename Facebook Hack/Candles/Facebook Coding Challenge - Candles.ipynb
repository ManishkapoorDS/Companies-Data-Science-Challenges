# Facebook Coding Challenge - Candles

```python
f = open("hard.txt", "r")
mylist = f.read().splitlines() 
```

```python
mylist = [i.split(' ') for i in mylist] 
```

```python
mylist[0]
```

```python
mylist[1]
```

## Easy

```python
list1 = [int(x) for x in mylist[1]]
list1.sort()
for i in list1:
    if int(1 + len(mylist[1]) -int(i)) > int(mylist[0][1]):
        res = int(mylist[1][len(mylist[1])-1]) - int(mylist[1][int(i)-1])
    else:
        res = int(mylist[1][int(i)-1])
        break

print(res)
```

## Medium

```python
list1 = [int(x) for x in mylist[1]]
list1.sort()
num = sum(list1)/int(mylist[0][1])
int(num)
```

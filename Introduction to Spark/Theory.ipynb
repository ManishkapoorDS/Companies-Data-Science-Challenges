## Batch Processing vs Stream Processing

Batch processing: used to process all the data in one go <br>
    1) Data at rest <br>
    2) High Latency <br>
    3) Concerned with volume and variety <br>
eg. calculate a monthly payroll summary

Stream processing: used to process data in real time <br>
    1) Data in motion <br>
    2) Low latency <br>
    3) Concerned with velocity <br>
eg. is a bank transaction fradulent?
